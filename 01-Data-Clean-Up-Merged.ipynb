```
import pandas as pd
```

```
# Load processed dataframes
df1 = pd.read_csv('/work/Processed Data/CloudResearch-1_24.08.12_processed.csv')
df2 = pd.read_csv('/work/Processed Data/CloudResearch-2_24.10.06_processed.csv')
df3 = pd.read_csv('/work/Processed Data/CloudResearch-3_24.11.11_processed.csv')

# Add Survey_Number column
df1["Survey_Number"] = 1
df2["Survey_Number"] = 2
df3["Survey_Number"] = 3

# Merge DataFrames vertically
df_merged = pd.concat([df1, df2, df3], axis=0, ignore_index=True)
df_merged
```

```
# Users who use chatbot in survey 2
user_identification_df2 = df2["participantId"].tolist()
users_df2 = df2["Usage"].tolist()
users_chatbot = [user for i, user in enumerate(user_identification_df2) if users_df2[i] == "Yes"]

# frequency of chatbot users
freq_users = len(users_chatbot)
percent_users = (freq_users/len(user_identification_df2))*100
print(freq_users, percent_users, users_chatbot)
```

```
# Users who started using chatbot in survey 2 (after survey 1)
users_starting_df2 = df2["Usage Starting"].tolist()
users_starting_chatbot = [user for i, user in enumerate(user_identification_df2) if users_starting_df2[i] == "Yes"]

# frequency of chatbot users
freq_start_users = len(users_starting_chatbot)
percent_start_users = (freq_start_users/len(user_identification_df2))*100
print(freq_start_users, percent_start_users, users_starting_chatbot)
```

```
# Users who stopped usage after survey 1
stop_use_df2 = df2["Usage Stopping"].tolist()
participant_id_df1 = df1["participantId"].tolist()  # IDs from survey 1
timeframe_use_df1 = df1["Usage Timeframe"].tolist()
session_length_df1 = df1["Session Length"].tolist()

# Create a mapping of participantId to (timeframe_use, session_length) from survey 1 (df1)
survey1_mapping = {participant_id_df1[i]: (timeframe_use_df1[i], session_length_df1[i]) 
                   for i in range(len(participant_id_df1))}

# Find participants who stopped using chatbot in df2
users_stopped = [user for i, user in enumerate(user_identification_df2) if stop_use_df2[i] == "Yes"]

# Find corresponding timeframe and session length for those who stopped (if they exist in df1)
stopped_users_info = {}

for user in users_stopped:
    if user in survey1_mapping:
        # If the participant exists in survey 1, get their info
        stopped_users_info[user] = survey1_mapping[user]
    else:
        # Handle cases where participant is not found in survey 1 (e.g., assign None or another default)
        stopped_users_info[user] = ("Not Found", "Not Found")

print(stopped_users_info)
freq_stopped_users = len(stopped_users_info)
print(freq_stopped_users)
```

```
def propagate_values_across_surveys(df, columns_to_propagate, id_column='participantId', survey_column='Survey_Number'):
    """
    Propagate values from Survey 1 to other surveys for specified columns.
    
    Parameters:
    -----------
    df : pandas DataFrame
        The input dataframe containing all survey data
    columns_to_propagate : list
        List of column names to propagate from Survey 1
    id_column : str, default='participantId'
        Name of the column containing participant IDs
    survey_column : str, default='Survey'
        Name of the column containing survey numbers
    
    Returns:
    --------
    pandas DataFrame
        DataFrame with propagated values
    """
    # Create a copy of the input dataframe
    df_processed = df.copy()
    
    # Get values from Survey 1
    survey1_values = df[df[survey_column] == 1][
        [id_column] + columns_to_propagate
    ].drop_duplicates()
    
    # For each participant, propagate their Survey 1 values to other surveys
    df_processed = df_processed.merge(
        survey1_values,
        on=id_column,
        how='left',
        suffixes=('', '_survey1')
    )
    
    # For each column to propagate, fill in missing values with Survey 1 values
    for col in columns_to_propagate:
        mask = df_processed[col].isna()
        df_processed.loc[mask, col] = df_processed.loc[mask, f'{col}_survey1']
        # Drop the temporary survey1 column
        df_processed = df_processed.drop(f'{col}_survey1', axis=1)
    
    return df_processed

columns_to_propagate = [
    'BFI Openness',
    'BFI Conscientious',
    'BFI Extraversion',
    'BFI Agreeableness',
    'BFI Neurotic', 'Housing', 'Living With', 'Area Type'
]
 
df_propagated = propagate_values_across_surveys(df_merged, columns_to_propagate)
df_propagated.drop(columns=['Unnamed: 0', 'index'], inplace=True)
df_propagated[df_propagated['Survey_Number'] == 3]
```

```
df_s1cluster = pd.read_csv('/work/Output Data/CloudResearch-1_24.08.12_clustered.csv')
cluster_mapping = df_s1cluster[['participantId', 'Cluster']].set_index('participantId')['Cluster']

# Add the s1_cluster column to df_propagated, leaving NaN for participants not in df_s1cluster
df_propagated['s1_cluster'] = df_propagated['participantId'].map(cluster_mapping)
```
